```python
#To find a summary of missing values in percentage
def missing_value_percentage_summary(df):
    total = df.shape[0]
    missing_columns = [col for col in df.columns if df[col].isnull().sum() > 0]
    missing_percent = {}
    for col in missing_columns:
        null_count = df[col].isnull().sum()
        percent = (null_count / total) * 100
        missing_percent[col] = percent
        print("{} : {} ({}%)".format(col, null_count, round(percent, 3)))
    return missing_percent
```

```python

# Fill the null values of categorical values
df['loan_limit'] = df['loan_limit'].fillna(df['loan_limit'].mode()[0])
df['approv_in_adv'] = df['approv_in_adv'].fillna(df['approv_in_adv'].mode()[0])
df['loan_purpose'] = df['loan_purpose'].fillna(df['loan_purpose'].mode()[0])
df['Neg_ammortization'] = df['Neg_ammortization'].fillna(df['Neg_ammortization'].mode()[0])
df['age'] = df['age'].fillna(df['age'].mode()[0])
df['submission_of_application'] = df['submission_of_application'].fillna(df['submission_of_application'].mode()[0])

# Fill the null values of numerical datatypes 
df['rate_of_interest'] = df['rate_of_interest'].fillna(df['rate_of_interest'].median())
df['Interest_rate_spread'] = df['Interest_rate_spread'].fillna(df['Interest_rate_spread'].median())
df['Upfront_charges'] = df['Upfront_charges'].fillna(df['Upfront_charges'].median())
df['term'] = df['term'].fillna(df['term'].mean())
df['property_value'] = df['property_value'].fillna(df['property_value'].median())
df['income'] = df['income'].fillna(df['income'].median())
df['LTV'] = df['LTV'].fillna(df['LTV'].median())
df['dtir1'] = df['dtir1'].fillna(df['dtir1'].mean())
```

# ML Algorithm

```python
# Split Independent and dependent features

X = df.drop(columns = ['Status'], axis = 1)
y = df['Status']
```

```python
# Importing necessary libraries for data preprocessing, model training, and evaluation

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier, StackingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25, random_state = 42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Logistic Regression

model1 = LogisticRegression(solver='saga', max_iter=1000)
model1.fit(X_train_scaled, y_train)
y_pred_model1 = model1.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred_model1)
print("Accuracy:", accuracy)
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
score = cross_val_score(model1, X_scaled, y, cv = 5)
print("Cross Validation scores: ", score)
```

```python
np.mean(score)*100
```

```python
#Decision Tree Classifier

model2 = DecisionTreeClassifier(random_state = 42)
model2.fit(X_train_scaled, y_train)
y_pred_model2 = model2.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred_model2)
print("Accuracy:", accuracy)
```

```python
score = cross_val_score(model2, X_scaled, y, cv = 5)
print("Cross Validation scores of Decision Tree : ", np.mean(score) * 100)
```

```python
# Randon Forest Classifier

model3 = RandomForestClassifier(random_state = 42)
model3.fit(X_train_scaled, y_train)
y_pred_model3 = model3.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred_model3)
print("Accuracy:", accuracy)
```

```python
score = cross_val_score(model3, X_scaled, y, cv = 5)
print("Cross Validation scores of Random Forest Classification : ", np.mean(score) * 100)
```

```python
# KNeighbor Model

model4 = KNeighborsClassifier(n_neighbors = 3)
model4.fit(X_train_scaled, y_train)
y_pred_model4 = model4.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred_model4)
print("Accuracy:", accuracy)
```

```python
score = cross_val_score(model4, X_scaled, y, cv = 5)
print("Cross Validation scores of KNeighbors Model : ", np.mean(score) * 100)
```

```python
from sklearn.metrics import classification_report

def generate_classification_report(model_name, y_test, y_pred):
    report = classification_report(y_test, y_pred)
    print(f"Classification Report For {model_name}:\n{report}\n")

generate_classification_report(model1, y_test, y_pred_model1)
generate_classification_report(model2, y_test, y_pred_model2)
generate_classification_report(model3, y_test, y_pred_model3)
generate_classification_report(model4, y_test, y_pred_model4)
```

```python
# Hybrid Model Using Voting Classifier

voting_model = VotingClassifier(estimators=[
    ('dt', model2),
    ('rf', model3)
], voting='soft')  # Use 'soft' for probabilities or 'hard' for class labels

voting_model.fit(X_train_scaled, y_train)
y_pred_voting = voting_model.predict(X_test_scaled)
accuracy_voting = accuracy_score(y_test, y_pred_voting)
print("Voting Classifier Accuracy:", accuracy_voting)
```

```python
voting_model2 = VotingClassifier(estimators=[
    ('dt', model2),
    ('lr', model1)  # Replacing Random Forest with Logistic Regression
], voting='soft')  # Use 'soft' for probabilities or 'hard' for class labels

voting_model2.fit(X_train_scaled, y_train)
y_pred_voting2 = voting_model2.predict(X_test_scaled)
accuracy_voting2 = accuracy_score(y_test, y_pred_voting2)
print("Voting Classifier Accuracy:", accuracy_voting2)
```

```python
voting_model3 = VotingClassifier(estimators=[
    ('kn', model4),
    ('lr', model1)  # Replacing Random Forest with kneighbour
], voting='soft')  # Use 'soft' for probabilities or 'hard' for class labels

voting_model3.fit(X_train_scaled, y_train)
y_pred_voting = voting_model3.predict(X_test_scaled)
accuracy_voting3 = accuracy_score(y_test, y_pred_voting)
print("Voting Classifier Accuracy:", accuracy_voting3)
```

```python
# Hybrid Model Using Stacking Classifier

stacking_model = StackingClassifier(estimators=[
    ('dt', model2),
    ('rf', model3)
], final_estimator=LogisticRegression())  # You can use any model here as final estimator

stacking_model.fit(X_train_scaled, y_train)
y_pred_stacking = stacking_model.predict(X_test_scaled)
accuracy_stacking = accuracy_score(y_test, y_pred_stacking)
print("Stacking Classifier Accuracy:", accuracy_stacking)
```

```python
stacking_model2 = StackingClassifier(estimators=[
    ('dt', model2),
    ('lr', model1)
], final_estimator=LogisticRegression())  # You can use any model here as final estimator

stacking_model2.fit(X_train_scaled, y_train)
y_pred_stacking2 = stacking_model2.predict(X_test_scaled)
accuracy_stacking2 = accuracy_score(y_test, y_pred_stacking2)
print("Stacking Classifier Accuracy:", accuracy_stacking2)
```

```python
stacking_model3 = StackingClassifier(estimators=[
    ('kn', model4),
    ('lr', model1)
], final_estimator=LogisticRegression())  # You can use any model here as final estimator

stacking_model3.fit(X_train_scaled, y_train)
y_pred_stacking3 = stacking_model3.predict(X_test_scaled)
accuracy_stacking3 = accuracy_score(y_test, y_pred_stacking3)
print("Stacking Classifier Accuracy:", accuracy_stacking3)
```
